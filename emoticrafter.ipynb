## 최근 출력물 확인

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# 결과 폴더의 파일 목록 확인
output_dir = '/content/drive/MyDrive/EmotiCrafter/outputs'
files = os.listdir(output_dir)

if files:
    # 가장 최근에 생성된 파일
    img_path = os.path.join(output_dir, files[0])
    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print("아직 생성된 파일이 없습니다.")
```
